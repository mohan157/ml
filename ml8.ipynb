```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

X = np.asarray([20, 5, 10, 13, 12])
Y = np.asarray([44, 22, 25, 32, 27])

b1 = 0
b0 = 0
L = 0.001
epochs = 1000
n = np.size(X)

for i in range(epochs):
    db1, db0, err, Total_Err_x, Total_Err_y = 0, 0, 0, 0, 0
    for x, y in zip(X, Y):
        y_pred = b1 * x + b0
        err = (y - y_pred)
        Total_Err_x = Total_Err_x + x * err
        Total_Err_y = Total_Err_y + err

    D_b1 = (-2 / n) * Total_Err_x
    D_b0 = (-2 / n) * Total_Err_y
    b1 = b1 - L * D_b1
    b0 = b0 - L * D_b0

print("b1:", b1, "b0:", b0)


Y_pred = b1 * X + b0
MSE = mean_squared_error(Y, Y_pred)
print("MSE:", MSE)

coefficient = r2_score(Y, Y_pred)
print("Coefficient of determination:", coefficient)

# Plotting
plt.figure(figsize=(10, 5))
plt.scatter(X, Y, label='Actual Data')
plt.plot(X, Y_pred, 'r', linewidth=2, label='Predicted')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
